```python
import pickle as pkl
import numpy as np

with open('single_X.pkl', 'rb') as file:
    X = np.array(pkl.load(file))

with open('single_y.pkl', 'rb') as file:
    y = np.array(pkl.load(file))

print(X.shape)
print(y.shape)
```

```python
from numpy.typing import NDArray


def flatten_last_dim(array: NDArray[np.float64]) -> NDArray[np.float64]:
    return np.reshape(array, (array.shape[0], array.shape[1], array.shape[2] * array.shape[3]))


X = flatten_last_dim(X)
print(X.shape)
print(y.shape)
```

```python

division_point = int(np.floor(X.shape[0] * 0.7))
x_train = X[:division_point]
y_train = y[:division_point]
x_test = X[division_point:]
y_test = y[division_point:]
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

import tensorflow as tf
from keras import Model
from keras.layers import Dense, Dropout, LSTM, Input, Lambda

inputs = Input(shape=(x_train.shape[1:]))
lstm1 = LSTM(512, input_shape=(x_train.shape[1:]), return_sequences=True, stateful=False)(inputs)
dropout1 = Dropout(0.2)(lstm1)
lstm2 = LSTM(512, stateful=False, return_sequences=True,)(dropout1)
dropout2 = Dropout(0.2)(lstm2)
lstm3 = LSTM(512, stateful=False, return_sequences=True,)(dropout2)
dropout3 = Dropout(0.2)(lstm3)
lstm4 = LSTM(512, stateful=False, return_sequences=True, )(dropout3)
dropout4 = Dropout(0.2)(lstm4)
lstm5 = LSTM(256, stateful=False, return_sequences=True)(dropout4)
last_hidden_state = Lambda(lambda x: x[:, -1, :])(lstm5)
dense2 = Dense(10, activation='softmax')(last_hidden_state)
model = Model(inputs=inputs, outputs=dense2)
opt = tf.keras.optimizers.Adam(learning_rate=1e-4)
hidden_state_ext_model = Model(inputs=inputs, outputs=lstm5)
model.compile(loss='sparse_categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test))
```

```python

c =hidden_state_ext_model.predict(X[0][None, :])
hidden_state_ext_model.save('hidden_state_ext.keras')
print(c.shape)
c
```
